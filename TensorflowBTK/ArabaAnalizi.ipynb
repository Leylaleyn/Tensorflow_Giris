```python
dataFrame = dataFrame.drop(["transmission","model","fuelType"],axis=1)  # transmission ve model bizim için gerekli değil o yuzden onu da çıkardık
```

```python
dataFrame.head()
```

```python
y = dataFrame["price"].values
x = dataFrame.drop("price",axis=1).values  # values numpy dizisine çeviriyor
```

```python
y
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=10)
```

```python
len(x_train)
```

```python
len(x_test)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
x_train = scaler.fit_transform(x_train)
```

```python
x_test = scaler.fit_transform(x_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
x_train.shape  # şöyle bir ipucu istersek katmanlarımızı 5 den başlatabiliriz
```

```python
model = Sequential()

model.add(Dense(12,activation="relu"))
model.add(Dense(12,activation="relu"))
model.add(Dense(12,activation="relu"))
model.add(Dense(12,activation="relu"))

model.add(Dense(1))

model.compile(optimizer="adam",loss="mse")
```

```python
# modeli oluşturduk şimdi eğiticez
```

```python
model.fit(x=x_train, y=y_train, validation_data=(x_test,y_test),batch_size=250,epochs=250)
```

```python
kayipVerisi = pd.DataFrame(model.history.history)
```

```python
kayipVerisi.head()
```

```python
kayipVerisi.plot()
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
tahminDizisi = model.predict(x_test)
```

```python
tahminDizisi
```

```python
mean_absolute_error(y_test,tahminDizisi)
```
